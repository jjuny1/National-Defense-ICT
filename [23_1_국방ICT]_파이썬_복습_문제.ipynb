## 참고 사항
- map() 함수 

  - 반복 가능한 객체(리스트, 튜플 등)의 요소를 지정된 함수로 처리해주는 함수
  - ```data = list(map(int, input().split()))```



````python
# 리스트 a에 포함된 원소들을 모두 정수로 변환 
 
a = [1.2, 2.5, 3.7, 4.6]
for i in range(len(a)):
  a[i] = int(a[i])

print(a)
````

````python
# 위와 같이 for 반복문을 사용하기 번거롭다. -> map을 사용하자. 

a = [1.2, 2.5, 3.7, 4.6]
a = list(map(int, a))

print(a)
````

## Q1. 사용자로 부터 리스트를 입력 받아, 원소들의 합을 출력하시오 
- 예) 입력 : [1, 2, 3, 4] -> 출력: 10

````python
data = list(map(int, input().split()))
## map() 리스트의 요소를 지정된 함수로 처리해주는 함수

sum = 0 
for i in data:
  sum = sum + i

print("출력:", sum)
````

## Q5. 사용자로 부터 숫자(n)를 입력받아, $n^{3}$이 큰지 $n!$이 큰지 판단하시오. 
- 입력: 1 -> 출력 : $n^{3}$과 $n!$이 같습니다. 
- 입력: 2 -> 출력 : $n^{3}$이 더 큽니다. 
- 입력: 6 -> 출력 : $6!$이 더 큽니다. 

````python
import math

n = int(input())

squared = n*n*n
fact = math.factorial(n)

if squared > fact :
  print("n^3이 더 큽니다")
elif squared < fact :
  print("n!이 더 큽니다")
else : 
  print("n!과 n^3이 같습니다.")


````

## Q6. 사용자로부터 리스트를 입력받아, 원소들을 크기 순으로 정렬하시오. 
- 입력: [1, 2, 3, 4] -> 출력: [1, 2, 3, 4]
- 입력: [1, -1, 3, -3] -> 출력: [-3, -1, 1, 3] 
- 입력: [1, 10, 20, -1, -2, -3] -> 출력: [-3, -2, -1, 1, 10, 20] 

* sort() 함수를 사용하지 말것

````python
arr = list(map(int, input().split()))
for i in range(len(arr) - 1, 0, -1):
  for j in range(i):
    if arr[j] > arr[j + 1]:
      arr[j], arr[j + 1] = arr[j + 1], arr[j]

print(arr)      
````

## Q6. 잔돈 돌려주기 프로그램을 만들어 보자. (자판기에 1000원을 지불한다고 가정). 예를 들어, 물건 가격이 400원이면 총 600원을 거슬러줘야하므로 500원 1개, 100원 1개, 50원 0개, 10원 0개를 거슬러줘야 한다. 거슬러주는 동전의 갯수가 최소가 되도록 하시오. 
- 입력 : 400 -> 출력 : [1, 1, 0, 0] 
- 입력 : 250 -> 출력 : [1, 2, 1, 0]
- 입력 : 980 -> 출력 : [0, 0, 0, 2] 

````python
n = int(input())

change = 1000 - n 
output = [0, 0, 0, 0]

while change > 0: 
  if change >= 500:
    change = change - 500
    output[0] = output[0] + 1
  elif change >= 100:
    change = change - 100
    output[1] = output[1] + 1
  elif change >= 50:
    change = change - 50
    output[2] = output[2] + 1
  elif change >= 10:
    change = change - 10
    output[3] = output[3] + 1
  else :
    break 

print(output)


````

## Q8. 아래 공식을 이용하여 파이값을 계산하는 함수 f를 작성하시오. 
- f(10), f(100), f(100)을 각각 출력하시오.

image.png

````python
def calculate_pi(n_terms):
  numerator : float = 4.0
  denominator : float = 1.0
  operation : float = 1.0
  pi : float = 0.0
  for _ in range(n_terms):
    pi += operation * (numerator/denominator)
    denominator += 2.0 # 2씩 증가 
    operation *= -1.0  # 빼기와 더하기 연산이 번갈아 반복 
  return pi 
````

## Q9. 두 팀으로부터 점수를 입력받아 첫번째 팀이 두번째 팀을 얼마나 많이 이겼는지를 계산하는 count_wins라는 함수를 작성하시오.

## 10. 짝궁 찾기 프로그램을 작성하시오. 
원소내에 동일한 원소의 쌍의 개수 출력하기 
- 입력: [1, 2, 3, 4, 1, 5, 5,] -> 출력: 2

````python
def count_pairs(arr):
    count = 0
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[i] == arr[j]:
                count += 1
                # Remove the matched element from the list to avoid counting it twice
                arr.pop(j)
                break
    return count

# 사용 예시
my_list = [1, 2, 3, 2, 1, 4, 5, 4]
print(count_pairs(my_list)) # 출력: 2
````

````python
## 4. Write a Python program to count the number of vowels in a string.
## 4번 문제 해답
def count_vowels(string):
    vowels = "aeiouAEIOU"
    count = 0
    for char in string:
        if char in vowels:
            count += 1
    return count

# example usage
string = "Hello, World!"
print(count_vowels(string)) # output: 3
````

````python
## 9. Write a Python program to check whether a given string is a palindrome or not.
## 9번 문제 해답
def is_palindrome(string):
    string = string.lower()
    reversed_string = string[::-1]
    return string == reversed_string

# example usage
string = "Racecar"
print(is_palindrome(string)) # output: True
````
