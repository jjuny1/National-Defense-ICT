### 1. 주어진 임의의 정수 리스트에 대하여 자연수의 합을 구하는 natural_sum 함수를 작성하세요.

* 예를 들어, 주어진 리스트 l = [1, 3, -2, 4] 인 경우 1 + 3 + 4 = 8 을 반환하게 됩니다.

힌트

* 반복문을 이용하여 리스트의 각 원소들을 확인합니다.
* 특정 변수에 각 원소의 합을 저장합니다.

```python
def natural_sum(l):

###### 아래 부터 코드를 작성하시오. ###############
    total = 0
    for num in l:
        if num > 0:
            total += num
    return total

## 아래 사항은 수정하지 마세요 ##
if __name__ == "__main__":
	l = eval(input())
	print(natural_sum(l))
```

* scikit-learn 라이브러리 사용 가능

### 문제 6-1. iris 데이터셋에 포함되어 있는 데이터의 수를 출력하는 코드를 작성하시오. (4점)

### 문제 6-2. 각 데이터의 sepal_length와 petal_length의 평균값을 출력하는 코드를 작성하시오.(4점)

### 문제 6-3. k-means 학습 종료 후 최종 centroid들에 대한 정보를 출력하시오. (6점)

### 문제 6-4. centroid의 ID가 각각 "0", "1", "2"라고 할 때, centroid "1"에 속하는 데이터들의 sepal_length 평균값을 출력하는 코드를 작성하시오. (6점)

### 문제 6-5.  matplotlib 라이브러리를 이용하여 클러스터링 결과를 나타내는 산포도를 출력하는 코드를 작성하시오. 이때, 클러스터 중심점을 검은색으로 표시하라.  (6점)

### 문제 6-6. 아래 꽃들은 몇 번째 클러스터에 속하는가?(4점)
* 첫번째 꽃 
  * sepal_leangth : 1
  * sepal_width : 1
  * petal_length : 1
  * petal_width : 1

* 두번째 꽃 
  * sepal_leangth : 6
  * sepal_width : 6
  * petal_length : 6
  * petal_width : 6



```python
# 아래 코드를 실행한 후 문제를 해결하시오. 

##############################################
############ 수정 금지 #####################

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()

############ 수정 금지 #####################
##############################################
```

```python
#### 답안 : 6-1 ####
#### 아래 부터 코드를 작성하시오. ########
print("데이터의 수:", len(iris.data))
```

```python
#### 답안 : 6-2 ####
#### 아래 부터 코드를 작성하시오. ########
import numpy as np
sepal_length_mean = np.mean(iris.data[:, 0])
petal_length_mean = np.mean(iris.data[:, 2])
print("sepal_length 평균값:", sepal_length_mean)
print("petal_length 평균값:", petal_length_mean)
```

```python
#### 답안 : 6-3 ####
#### 아래 부터 코드를 작성하시오. ########
import warnings
warnings.filterwarnings('ignore')

X = iris.data[:,[0, 2]]
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
print("최종 centroid 정보:\n", kmeans.cluster_centers_)
```

```python
#### 답안 : 6-4 ####
#### 아래 부터 코드를 작성하시오. ########
centroid_1_data = X[kmeans.labels_ == 1]
centroid_1_sepal_length_mean = np.mean(centroid_1_data[:, 0])
print("centroid 1에 속하는 데이터들의 sepal_length 평균값:", centroid_1_sepal_length_mean)
```

```python
#### 답안 : 6-5 ####
#### 아래 부터 코드를 작성하시오. ########
plt.scatter(X[kmeans.labels_ == 0, 0], X[kmeans.labels_ == 0, 1], c='red', label='Cluster 0')
plt.scatter(X[kmeans.labels_ == 1, 0], X[kmeans.labels_ == 1, 1], c='blue', label='Cluster 1')
plt.scatter(X[kmeans.labels_ == 2, 0], X[kmeans.labels_ == 2, 1], c='green', label='Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='black', marker='x', label='Centroids')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()
plt.show()
```

```python
#### 답안 : 6-6 ####
#### 아래 부터 코드를 작성하시오. ########

flower_1 = np.array([[1, 1]])
flower_2 = np.array([[6, 6]])

cluster_1 = kmeans.predict(flower_1)
cluster_2 = kmeans.predict(flower_2)

print("첫번째 꽃이 속한 클러스터:", cluster_1[0])
print("두번째 꽃이 속한 클러스터:", cluster_2[0])
```

```python
# 아래 코드를 실행한 후 문제를 해결하시오. 

##############################################
############ 수정 금지 #####################
##############################################
import random 
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

scores = [(random.randint(20, 100), random.randint(30, 99) ) for x in range(1000)]
grades = [random.choice(['A']) if np.mean(scores[i]) > 85 else random.choice(['B', 'C']) if np.mean(scores[i]) > 70  else random.choice(['C']) if np.mean(scores[i]) > 60  else random.choice(['E']) for i in range(1000)]
##############################################



```

```python
#### 답안 : 8-1 ####
#### 아래 부터 답안 코드를 작성하시오.(각 문항별로 답안을 작성하시오) ########
# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(scores, grades, test_size=0.3, random_state=42)

# 모델 학습
knn = KNeighborsClassifier(n_neighbors=7, weights='distance')
knn.fit(X_train, y_train)

# 모델 평가
accuracy = knn.score(X_test, y_test)
print(f"정확도 : {accuracy:.3f}")
```

```python
#### 답안 : 8-2 ####
#### 아래 부터 코드를 작성하시오.(각 문항별로 답안을 작성하시오) ########
# 이육사 생도의 점수
x_yuksa = [[78, 57]]

# 예측
y_yuksa = knn.predict(x_yuksa)

print(f"예측 학점 : {y_yuksa[0]}")
```
