```python
############이 아래 코드는 실행만 하고, 수정하지 마시오############
from sklearn.datasets import load_breast_cancer  #sklearn 라이브러리에서 특정 함수를 import
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score  #confusion_matrix 등 다양한 성능평가 함수 존재 
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt #matplotlib 라이브러리에서 pyplot 함수를 plt로 호출

sns.set()
breast_cancer = load_breast_cancer()
X = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names) # 데이터 프레임 형태 선언
X
```

- 5.(c) 답: 최대 정확도를 갖는 K 값 = (    )

```python
############6.(a) - print() 함수 등을 이용하여, k값에 따른 정확도를 출력하시오.############
#X = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names) # 데이터 프레임 형태 선언

X = X[['mean texture', 'mean area', 'mean symmetry']] # 세 개 열만 X로 저장
y = pd.Categorical.from_codes(breast_cancer.target, breast_cancer.target_names)
y = pd.get_dummies(y, drop_first=True)  # Convert categorical variable into dummy/indicator variables. 


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1) #학습용 데이터와 테스트용 데이터를 분리

blank = []
for k in range (1,40):
  knn = KNeighborsClassifier(n_neighbors=k, metric= 'euclidean')
  knn.fit(X_train, y_train.values.ravel())
  y_pred = knn.predict(X_test)
  from sklearn.metrics import accuracy_score # confusion_matrix 등 다양한 성능평가 함수 존재 
  import numpy as np

  print('k값: ',k, ' 정확도---->', accuracy_score(y_test, y_pred))
  blank.append(accuracy_score(y_test, y_pred))

```

```python
############6.(b) - pyplot 함수 등을 이용하여, k값에 따른 정확도를 그래프로 표현 하시오.############
import matplotlib.pyplot as plt
x1 = np.arange(39)+1
y1 = np.array(blank)
plt.plot(x1,y1)
plt.show()
```
