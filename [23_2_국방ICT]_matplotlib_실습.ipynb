## 예제

```python
# 1. Create a line graph to show the number of students enrolled in a university over the past 5 years.
import matplotlib.pyplot as plt

years = [2017, 2018, 2019, 2020, 2021]
enrollment = [10000, 12000, 15000, 18000, 20000]

plt.plot(years, enrollment)
plt.title('University Enrollment')
plt.xlabel('Year')
plt.ylabel('Number of Students')
plt.show()
```

```python
# 2. Create a bar chart to show the number of students in each major at a university.
import matplotlib.pyplot as plt

majors = ['Computer Science', 'Biology', 'Psychology', 'Business', 'Engineering']
students = [800, 600, 500, 400, 300]

plt.bar(majors, students)
plt.title('University Majors')
plt.xlabel('Major')
plt.ylabel('Number of Students')
plt.show()
```

```python
# 3. Create a scatter plot to show the relationship between the hours studied and the exam scores of a group of students.
import matplotlib.pyplot as plt

hours_studied = [4, 5, 2, 6, 7, 8, 3, 1, 9, 10]
exam_scores = [75, 85, 60, 90, 95, 98, 65, 50, 100, 90]

plt.scatter(hours_studied, exam_scores)
plt.title('Study Hours vs Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```

```python
# 4. Create a pie chart to show the percentage of students from different countries in a university.
import matplotlib.pyplot as plt

countries = ['USA', 'China', 'India', 'Brazil', 'Mexico']
students = [5000, 3000, 2000, 1500, 1000]

plt.pie(students, labels=countries, autopct='%1.1f%%')
plt.title('University Student Nationalities')
plt.show()
```

```python
# 5. Create a histogram to show the distribution of grades in a class.
import matplotlib.pyplot as plt

grades = [75, 85, 60, 90, 95, 98, 65, 50, 100, 90, 80, 75, 85, 70, 90, 95, 75, 80, 90, 95]

plt.hist(grades, bins=10, edgecolor='black')
plt.title('Class Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.show()
```
